```python
import pandas as pd
```

```python
gdp_data = pd.read_html('http://statisticstimes.com/economy/gdp-indicators-2018.php')
```

```python
gdp_dataset = gdp_data[1]
```

```python
gdp_dataset
```

```python
gdp_dataset.info()
```

```python
gdp_dataset['GDP_BUSD'] = gdp_dataset['GDP 2018 (billions of $)']['Nominal']
```

```python
gdp_dataset['GDP_per_capita_USD'] = gdp_dataset['GDP per capita 2018 ($)']['Nominal']
```

```python
gdp_dataset['country'] = gdp_dataset['Country/Economy']['Country/Economy']
```

```python
columns = ['country','GDP_BUSD', 'GDP_per_capita_USD']
```

```python
gdp_dataset_cleaned = gdp_dataset[columns]
gdp_dataset_cleaned.columns = gdp_dataset_cleaned.columns.droplevel(1)
```

```python
def clean_gdp(gdp):
    if gdp != '-':
        return gdp
    else:
        return None
```

```python
gdp_dataset_cleaned['GDP_BUSD'] = gdp_dataset_cleaned['GDP_BUSD'].apply(clean_gdp)
```

```python
gdp_dataset_cleaned['GDP_per_capita_USD'] = gdp_dataset_cleaned['GDP_per_capita_USD'].apply(clean_gdp)
```

```python
gdp_dataset_cleaned.dropna(subset=['GDP_BUSD'], inplace=True)
```

```python
gdp_dataset_cleaned.dropna(subset=['GDP_per_capita_USD'], inplace=True)
```

```python
gdp_dataset_cleaned
```

```python
df_forbes = pd.read_parquet(f'../data/processed/cleaned_data.parquet')

df_forbes.head()
```

```python
#The link below contains the 2018 Forbes List with the country column totally full.
forbes_url = 'https://www.forbes.com/ajax/list/data?year=2018&uri=billionaires&type=person'
df_country = pd.read_json(forbes_url, orient='records')
df_country['gender'].unique()
```

```python
null_cols = df_country.isnull().sum()

null_cols[null_cols > 0]

null_cols[null_cols > 0] / len(df_country) * 100
```

```python
columns = ['name', 'country', 'gender']
df_country_to_merge = df_country[columns]
```

```python
df_country_to_merge.head()
```

```python
df_forbes_country = pd.merge(df_forbes, df_country_to_merge, on='name')
df_forbes_country
```

```python
df_forbes_country['country'] = df_forbes_country['country_y']
columns = ['name', 'worth_amount_(BUSD)', 'country', 'company_sector', 'company_name', 'age']
df_forbes_country_cleaned = df_forbes_country[columns]
df_forbes_country_cleaned
```

```python
null_cols = df_forbes_country_cleaned.isnull().sum()

null_cols[null_cols > 0]

null_cols[null_cols > 0] / len(df_forbes_country_cleaned) * 100
```

```python
df_forbes_country_gdp = pd.merge(df_forbes_country_cleaned, gdp_dataset_cleaned, on='country')
df_forbes_country_gdp.head()
```

```python
df_forbes_country_gdp.to_parquet(f'../data/processed/forbeslist_gdp.parquet')
```
